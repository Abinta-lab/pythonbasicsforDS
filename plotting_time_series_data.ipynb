**Time series data**, also referred to as time-stamped data, is a sequence of data points indexed in time order.

Time series data is a collection of observations obtained through repeated measurements over time. Plot the points on a graph, and one of your axes would always be time.

Examples of time series analysis:
1.   Electrical activity in the brain
2.   Rainfall measurements
3.   Stock prices
4.   Number of sunspots
5.   Annual retail sales
6.   Monthly subscribers
7.   Heartbeats per minute

```python
from matplotlib.lines import lineStyles
import pandas as pd
from datetime import datetime, timedelta  # The datetime module supplies classes for manipulating dates and times.
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates

plt.style.use('seaborn')

dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates, y, linestyle = 'solid')

# Rotate and align the tick labels so they look better.
plt.gcf().autofmt_xdate()

date_format = mpl_dates.DateFormatter('%b %d, %Y') #%m %d, %Y')

#plt.gca() used to get the current Axes instance on the current figure matching the given keyword args, or create one. 
plt.gca().xaxis.set_major_formatter(date_format)


plt.tight_layout()

plt.show()
```

```python
import pandas as pd
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates

plt.style.use('seaborn')

data = pd.read_csv('/content/gdrive/MyDrive/DS211/NoteBooks/Matplotlib/data_time.csv')

data['Date'] = pd.to_datetime(data['Date'])
data.sort_values('Date', inplace=True)

price_date = data['Date']
price_close = data['Close']

plt.plot_date(price_date, price_close, linestyle='solid')

plt.gcf().autofmt_xdate()

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()

plt.show()
```
