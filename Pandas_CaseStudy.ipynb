### IPL EDA

You have to find out top five batsmans in terms of runs scored in all the seasons of IPL!

```python
delivery = pd.read_csv('/content/gdrive/MyDrive/DS211/NoteBooks/deliveries.csv')
delivery.head(10)
```

```python
group_batsman = delivery.groupby('batsman')
```

```python
group_batsman['batsman_runs'].sum().sort_values(ascending=False).head()
```

```python
# Find the batsman who hitted largest number of six in all the seasons.
mask =  delivery['batsman_runs']==6
new_delivery = delivery[mask]
new_delivery.head()
```

```python
max_six = new_delivery.groupby('batsman')
```

```python
max_six['batsman_runs'].count().sort_values(ascending=False).head()
```

```python
# Lets create a function in which you pass name of a batsman as a string it will return you against which team he scored the highest runs. 
def run_scored(batsman_name):
    mask =  delivery['batsman']== batsman_name
    only_kohli = delivery[mask]
    team = only_kohli.groupby('bowling_team')
    return team['batsman_runs'].sum().sort_values(ascending=False).head(3).index[0] 
     
```

```python
run_scored('V Kohli')
```
